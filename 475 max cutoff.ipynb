```python
import librosa
import numpy as np 
import zipfile
import os


maxilist = []
with zipfile.ZipFile("new475mus.zip", "r") as f:
# Approximate maximum frequencies with roll_percent=0.85 (default)
    for name in f.namelist():
        #data = f.read(name)
        y, sr = librosa.load(name)
        # Approximate maximum frequencies with roll_percent=0.85 (default)
        rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        maxi = 0
        for roll in rolloff:
            for sing in roll:
                if sing > maxi:
                    maxi = sing
            maxilist.append(maxi)
            print (maxilist)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
y = [0]*len(maxilist)
plt.plot(maxilist, y, 'o', color='black');
plt.show()
print (maxilist)
```
